```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delays_us = np.array([8646,47434,92576,141286,188523,283536,293302,283065,338618,1823754])
sizes_B = np.array([100,500,1000,1500,2000,3000,3100,3150,3250,3300])

delays_s = delays_us * (10**(-6))
sizes_b = sizes_B * 8
print(delays_s)
print(sizes_b)
```

```python
plt.plot(sizes_B, delays_s, color = "orange")
plt.plot(sizes_B, delays_s, "ro")
plt.grid(True)
plt.title("Packet Size vs Delay")
plt.ylabel("Delay (seconds)")
plt.xlabel("Size (bytes)")
plt.savefig("delays")
plt.show()
```

```python
throughput = sizes_b / delays_s /10**3
print(throughput)
```

```python
plt.plot(sizes_b, throughput, color="purple")
plt.grid(True)
plt.title("Packet Size vs Throughput")
plt.ylabel("Throughput (kb/s)")
plt.xlabel("Size (bits)")
plt.savefig("throughputs")
plt.show()
```
